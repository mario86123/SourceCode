```python
import matplotlib.pyplot as plt
```

```python
path = 'SDF_best_fitness/LLMST/20_100_0.txt'
problem = path.split('/')[0][0:3]
model = path.split('/')[1]
file_name = path.split('/')[2]
ell = int(file_name.split('_')[0])
pop_size = int(file_name.split('_')[1])




tmp = open(path, 'r').read().split()

NFE = []
best_fitness = []

for i in range(len(tmp)):
    if i % 2 == 0:
        NFE.append(int(tmp[i]))
    else:
        best_fitness.append(int(tmp[i]))

print(len(NFE))
print(len(best_fitness))

plt.plot(NFE, best_fitness)
            
# plt.legend()
plt.rcParams["figure.figsize"] = (10,10)
plt.title(f'best fitness vs. NFE:')
plt.xlabel ('NFE')
plt.ylabel ('best fitness')
# plt.ylim(ymax=1.05)  # this line
# plt.ylim(ymin=-0.05)  # this line
plt.show()
```
