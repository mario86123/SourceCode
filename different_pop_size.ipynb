```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_bar_chart(problem_name, arpd_lst):
    model = ['MST', 'NO', 'GMC', 'ET5']
#     ARPD = [0.044, 0.038, 0.131, 0.074, 0.053]

    x = np.arange(len(model))

    plt.bar(x, arpd_lst, color=['red', 'orange', 'gold', 'green'])
#     plt.bar(x, arpd_lst, color=['red', 'orange', 'gold', 'green', 'cornflowerblue', 'blue', 'purple'])
    
    for index,data in enumerate(arpd_lst):
        plt.text(x=index-0.5 , y =data+0.001 , s=f"{data}" , fontdict=dict(fontsize=25))

    plt.title(problem_name, fontsize=30)    

    plt.xticks(x, model, fontsize=20)
    # plt.xlabel('model', fontsize=20)
    
    plt.yticks(fontsize=15)
    plt.ylabel('ARPD', fontsize=20)

    plt.rcParams["figure.figsize"] = (8,8)
    plt.show()
```

```python
plot_bar_chart('SDF', [0.0, 17.3, 41.7, 57.0])
```

```python
plot_bar_chart('TSP', [25.8, 33.1, 76.9, 0.0])
```
