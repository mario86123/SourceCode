```python
import matplotlib.pyplot as plt
```

```python
def plot_fitness_NFE(path_lst):
    
    for path in path_lst:
    
#         path = 'SDF_best_fitness/NO/20_200_0.txt'
        problem = path.split('/')[0].split('_')[0]
        best_or_avg = path.split('/')[0].split('_')[1]
#         print(problem + ' ' + best_or_avg)
        model = path.split('/')[1]
        file_name = path.split('/')[2]

        ell = int(file_name.split('_')[0])
        pop_size = int(file_name.split('_')[1])




        tmp = open(path, 'r').read().split()

        NFE = []
        fitness = []

        for i in range(len(tmp)):
            if i % 2 == 0:
                NFE.append(int(tmp[i]))
            else:
                fitness.append(float(tmp[i]))

#         print(len(NFE))
#         print(len(fitness))


        plt.plot(NFE, fitness, label=model+' '+best_or_avg)

    plt.legend(fontsize=15)
    plt.rcParams["figure.figsize"] = (10,10)
    plt.title(f'{problem} {best_or_avg} fitness vs. NFE:', fontsize=30)
    plt.xlabel ('NFE', fontsize=20)
    plt.ylabel ('fitness', fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    # plt.ylim(ymax=1.05)  # this line
    # plt.ylim(ymin=-0.05)  # this line
#     plt.set_yscale('log')
#     plt.grid(which='major', alpha=0.5)
#     plt.grid(b=True, which='minor', linestyle='--', alpha=0.2)
#     plt.minorticks_on()
    plt.show()
```

```python
path_lst = ['SDF_best_fitness/LLMST59/60_2400_0.txt', 
            'SDF_best_fitness/GM/60_10_0.txt',
            'SDF_best_fitness/NO/60_2400_0.txt',
            'SDF_best_fitness/ET/60_10_0.txt']

plot_fitness_NFE(path_lst)
```
