```python
path = 'QAP_entropy/LLMST99/100a_100_LLMST99_1.txt'
tmp = open(path, 'r').read().split('\n\n')
tmp[1]
```

```python
for i in range(len(tmp)):
    tmp[i] = tmp[i].splitlines()
    print(tmp[i])
    print('\n')
```

```python
edge_num = int(tmp[0][0].split()[0])
edge_entropy = float(tmp[0][0].split()[1])

print(edge_num)
print(edge_entropy)
```

```python
ell = 100
already_sampled_node = edge_num % ell
not_sampled_yet_node = int((edge_num - already_sampled_node) / ell - 1)
print(already_sampled_node)
print(not_sampled_yet_node)
```
